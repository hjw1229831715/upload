```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('datasets/movielens/movies.dat', sep='::',header=None, names=mnames)
movies[:10]
```

```python
all_genres = []
for x in movies.genres: all_genres.extend(x.split('|'))
genres = pd.unique(all_genres)
genres
```

```python
zero_matrix = np.zeros((len(movies), len(genres)))
dummies = pd.DataFrame(zero_matrix, columns=genres)
gen = movies.genres[0]
gen.split('|')
```

```python
dummies.columns.get_indexer(gen.split('|'))  #获得gen.split里的三个分类在dummies.colimns的位置
```

```python
for i, gen in enumerate(movies.genres):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, indices] = 1
dummies
```

```python
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.iloc[0]
```

```python
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com
"""
pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'
regex = re.compile(pattern, flags=re.IGNORECASE)
regex.findall(text)
```

```python
m = regex.search(text)
m
```

```python
text[m.start():m.end()]
```

```python
print(regex.match(text)) #regex.match returns None, as it only will match if the pattern occurs at the start of the string
```

```python
print(regex.sub('REDACTED', text)) #替换
```

```python
pattern = r'([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})'
regex = re.compile(pattern, flags=re.IGNORECASE)
m = regex.match('wesm@bright.net')
m.groups()
```

```python
regex.findall(text)
```

```python
print(regex.sub(r'Username: \1, Domain: \2, Suffix: \3', text)) #1 2 3表示匹配的第一第二第三部分
```
